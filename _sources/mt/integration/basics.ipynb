# Defining Integration

In many of your mathematics courses to date, you have probably been exposed to integrals (especially, the popular Riemann integral). Here, we're going to build up a more generic definition of an integral; that is, the ability to perform integration on *any* measurable space, rather than just Riemann ones. The way we're going to do this is rather straightforward, and likely matches how your real analysis class built the Riemann integral: we're going to start with the most basic of functions (*simple* functions), and demonstrate that these simple definitions obey desirable and intuitive rules for the integral to have. Then, we're going to gradually relax this extremely simple operation to more and more complicated sorts of functions, until we have general rules for integration that will be sufficient for our treatment of probability theory.

## Simple functions

To build up integration rules, one of the most basic ways to do so is with something called simple functions:

````{prf:definition} Simple function
:label: mt:int:basics:simplefn
Suppose $(\Omega, \mathcal F)$ and $(\mathbb C, \Sigma)$ are measurable spaces, where $\mu$ is a measure on $(\Omega, \mathcal F)$. A function $\varphi \in m(\mathcal F, \Sigma)$ is said to be **simple** if, for disjoint sets $\{F_n\}_{n \in [N]}$ are s.t. $\mu(F_n) < \infty$ (that is, a $\mu$-finite sequence of disjoint measurable sets), then for all $\omega \in \Omega$:
```{math}
    \varphi(\omega) = \sum_{n \in [N]}\alpha_n \mathbb 1_{\{F_n\}}(\omega)
```
for a sequence $(\alpha_n)_{n \in [N]} \subseteq \mathbb C$.
````

Remember that since $\varphi \in m(\mathcal F, \Sigma)$, then for any $E \in \Sigma$, $\varphi^{-1}(E) = \{\omega \in \Omega : \varphi(\omega) \in E\} \in \mathcal F$. The function need not have a unique representation to be simple; that is, the sets, and the coefficients, don't necessarily have to be distinct. Let's see an example:

````{prf:example} Simple function is not unique
Suppose that $\Omega = [0, 1]$, and $F_1 = [0, 0.5)$ and $F_2 = [0.5, 1]$, and the measure $\mu$ is the Lebesgue measure. note that $F_1$ and $F_2$ are a partition of $\Omega$, and therefore disjoint, where $\lambda(F_n) = 0.5 < \infty$. If $\alpha_1 = 1$ and $\alpha_2 = 2$, we could further split the disjoint sets into $F_1' = [0, 0.25)$, $F_2 = [0.25, 0.5)$, $F_3' = [0.5, 0.75)$, and $F_4' = [0.75, 1]$, where $\alpha_1' = \alpha_2' = 1$, and $\alpha_3' = \alpha_4' 2$. Here, $\lambda(F_n') = 0.25 < \infty$. If $\varphi(\omega) = \sum_{n \in [N]} \alpha_n \mathbb 1_{F_n}(\omega)$ and $\varphi(\omega) = \sum_{n \in [N]} \alpha_n' \mathbb 1_{F_n'}(\omega)$, we can see that the representation is not unique, because $\varphi(\omega) = \varphi'(\omega)$ for all $\omega \in \Omega$. 
````

We start by defining the integral of a simple function, rather intuitively:

````{prf:definition}
:label: mt:int:basics:int:simplefn
Suppose that $\varphi : (\Omega, \mathcal F) \rightarrow (\mathbb C, \Sigma)$ is a simple function, with corresponding coefficients/disjoint sets $\{(\alpha_n, F_n)\}_{n \in [N]} \subseteq \mathbb C \times \Sigma$. We define the integral:
```{math}
    \int \varphi \,\text d \mu \triangleq \sum_{n \in [N]}\alpha_n \mu(F_n).
```
````
Conceptually, what is going on here is, we define the integral as an association with indicator functions, which we briefly touched on in the last chapter. The idea is that the "integral" of a single disjoint set is the "area" (measure) of the amount of "space" that each $F_n$ occupies in $\Omega$, and then we weight that by the corresponding coefficient $\alpha_n$. 

### Properties of the integral for simple functions

To show that the integral holds desirable properties, we need to extend a concept we defined previously: statements $\mathcal S$ that are $\mu$-a.e. Remember that we defined this term in {prf:ref}`mt:intro:prob_spaces:meas:ae`. Recalling that definition, we can now extend this concept to a very particular set of statements, *relations*:

````{prf:definition} Relation holds almost everywhere
:label: mt:int:basics:relae
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(S, \Sigma)$ is a measurable space. Assume that $\varphi, \psi : (\Omega, \mathcal F) \rightarrow (S, \Sigma)$ are two functions. We say that a relation $\rho : S \times S \rightarrow \{0, 1\}$ holds $\mu$-a.e. if there exists $N \in \mathcal F$ s.t.:
```{math}
    \mu\left(\{\omega \in N : \rho\left(\varphi(\omega), \psi(\omega)\right) = 0\}\right) = 0
```
and for every $\omega \in \Omega \setminus N$, $\rho\left(\varphi(\omega), \psi(\omega)\right) = 1$.

and we write that $\rho(\varphi, \psi)$ is $\mu$-a.e. If the measure is implied, we shorten this to a.e.
````

The idea here is that for some subset of the $\sigma$-algebra of the domain, the relation does *not* hold, but that the measure of this set of points where the relation does not hold has measure $0$. For every other point in the domain, the relation holds.

For instance, let's take an example to be $\rho(a, b) = a \geq b$ (a function which is true and returns $1$ when $a \geq b$, and false or returns $0$ when $a < b$). Then $\varphi \overset{a.e.}{\geq} \psi$ means that $\mu\left(\left\{\omega \in \Omega : \varphi(\omega) < \psi(\omega)\right\}\right) = 0$. Let's see how we can use this with the integral.

Intuitively, non-negative simple functions have non-negative integrals:

````{prf:property} Non-negative simple function has non-negative integral
:label: mt:int:basics:int:prop:nnintegral
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. Suppose that $\varphi : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ is a simple function. Then if $\varphi \overset{a.e.}{\geq} 0$, $\int \varphi\,\text d \mu \geq 0$. 
````
````{prf:proof}
If $\varphi \overset{a.e.}{\geq} 0$ is simple, let:
```{math}
    \varphi(\omega) = \sum_{n \in [N]} \alpha_n \mathbb 1_{\{F_n\}}(\omega)
```
be a (potentially non-unique) representation of it, which exists by {prf:ref}`mt:int:basics:simplefn` as $\varphi$ is simple.

Since $\varphi \overset{a.e.}{\geq} 0$, then $\alpha_n \geq 0$ for every $n$ except a subset of indexing sets $\mathcal M \subseteq [N]$, where for $m \in \mathcal M$, $\mu(F_m) = 0$. Then:

```{math}
    \int \varphi \,\text d \mu &= \sum_{n \in [N] \setminus \mathcal M}\alpha_n \mu(F_n) + \sum_{n \in \mathcal M} \alpha_n \mu(F_n) \\
    &= \sum_{n \in [N] \setminus \mathcal M}\alpha_n \mu(F_n) + 0\sum_{n \in \mathcal M}\alpha_n,\,\,\,\,\mu(F_n) = 0\text{ for }n \in \mathcal M \\
    &\geq 0.
```
````

We can rescale simple functions by constants, and the integral of the rescaled function is just the rescaled integral:

````{prf:property} Rescaling simple function has rescaled integral
:label: mt:int:basics:int:prop:rescale
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. Suppose that $\varphi : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ is a simple function. For any $r \in \mathbb R$, $\int r\varphi \,\text d \mu = r\int \varphi \,\text d\mu$.
````
````{prf:proof}
If $\varphi$ is simple and $r \in \mathbb R$, then:
```{math}
    r\varphi(\omega) = \sum_{n \in [N]} r\alpha_n \mathbb 1_{\{F_n\}}(\omega)
```
So $r\varphi$ is also simple with coefficients $\{r\alpha_n\}_{n \in [N]}$.

Then:
```{math}
    \int r\varphi \,\text d\mu &= \sum_{n \in [N]}r\alpha_n \mu(F_n) \\
    &= r\sum_{n \in [N]}\alpha_n \mu(F_n) \\
    &= r\int \varphi \,\text d\mu.
```
````

Further, we can add simple functions and the integral of their sum is the sum of their integrals. This sounds rather trivial, but due to a slight amount of nuance, we need a new definition first. Notice that the subsets, $\{F_n\}_{n \in [N]}$ of {prf:ref}`mt:int:basics:simplefn` *need not* necessarily be a partition of $\Omega$; we might be left with a set of points $\Omega \setminus \bigsqcup_{n \in [N]} F_n$ where $\omega \in N \Rightarrow \varphi(\omega) = 0$. This is the concept underlying the *support* of a function:

````{prf:definition} Support of a function
:label: mt:int:basics:int:prop:support
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb C, \Sigma)$ is a measurable space, and $f : (\Omega, \mathcal F) \rightarrow (\mathbb C, \Sigma)$ is a function. Then the support of $f$ is:
```{math}
    \text{support}(f) \triangleq \left\{\omega \in \Omega : f(\omega) \neq 0\right\}
```
````


Now, we can address the problem of integrating the sum of simple functions. To do so, we need a quick lemma: the sum of simple functions is simple:

````{prf:lemma} Sum of simple functions is simple
:label: mt:int:basics:int:prop:sum_simple_fn_simple
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. Suppose that $\varphi, \psi : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ are simple functions. Then $\varphi + \psi$ is simple.
````

````{prf:proof}
Since $\varphi, \psi$ are simple, then:
```{math}
    \varphi(\omega) = \sum_{n = 1}^N \alpha_n \mathbb 1_{\{A_n\}}(\omega) \text{ and } \psi(\omega) = \sum_{m = 1}^M \beta_m \mathbb 1_{\{B_m\}}(\omega).
```
We need to show that their sum is a simple function. 

Problematically, $\varphi$ and $\psi$ might not have the same *support*; that is, $\bigsqcup_{n \in [N]} A_n$ might not necessarily be $\bigsqcup_{m \in [M]} B_m$.

Let $A_0  \bigsqcup_{m \in [M]} B_m \setminus \bigsqcup_{n \in [N]}A_n$, and $B_0 = \bigsqcup_{n \in [N]} A_n \setminus \bigsqcup_{m \in [M]}B_m$. Here, $A_0$ is the set of all elements in $\bigsqcup_{m \in [M]} B_m$ that are not in $\bigsqcup_{n \in [N]} A_n$, and vice-versa. Note now that $\bigsqcup_{m = 0}^M B_m = \bigsqcup_{n = 0}^N A_n$.

Since $\{B_m\}_{m = 0}^M$ are still disjoint, then for every $\omega \in A_n$, there exists one $m \in \{0, ..., M\}$ where $\omega \in B_m$ (and only one, since $B_m$ are disjoint). Therefore, $A_n = \bigsqcup_{m = 0}^M A_n \cap B_m$, and $\mathbb 1_{\{A_n\}}(\omega) = \sum_{m = 0}^M \mathbb 1_{\{A_n \cap B_m\}}(\omega)$.

So:
```{math}
    \varphi(\omega) = \sum_{n = 1}^N \alpha_n \sum_{m = 0}^M \mathbb 1_{\{A_n \cap B_m\}}(\omega)
```
and vice-versa for $\psi(\omega)$. 

Letting $\alpha_0 = \beta_0 = 0$, then:
```{math}
    \varphi(\omega) + \psi(\omega) = \sum_{n = 0}^N \sum_{m = 0}^M (\alpha_n + \beta_m) \mathbb 1_{\{A_n \cap B_m\}}(\omega)
```
By construction, since $\{A_n\}_n$ are mutually disjoint and $\{B_m\}_m$ are mutually disjoint, that $\{A_n \cap B_m\}_{n, m}$ are also mutually disjoint, so $\varphi(\omega) + \psi(\omega)$ is simple.
````

and we are ready to address the quantity of interest:

````{prf:property} Summing simple functions have summed integral
:label: mt:int:basics:int:prop:sum_int
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. Suppose that $\varphi, \psi : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ are simple functions. Then $\int(\varphi : \psi) \,\text d \mu = \int \varphi \,\text d \mu + \int \psi \,\text d \mu$.
````

````{prf:proof}
Let $A_0, B_0, \alpha_0, \beta_0$ be defined as in {prf:ref}`mt:int:basics:int:prop:sum_simple_fn_simple`. Then:
```{math}
    \int \varphi + \psi \,\text d \mu &= \sum_{n = 0}^N \sum_{m = 0}^M (\alpha_n + \beta_m) \mu(A_n \cap B_m) \\
    &= \sum_{n = 0}^N \sum_{m = 0}^M \alpha_n \mu(A_n \cap B_m) + \sum_{n = 0}^N \sum_{m = 0}^M \beta_m \mu(A_n \cap B_m) \\
    &= \sum_{n = 0}^N \alpha_n \sum_{m = 0}^M \mu(A_n \cap B_m) + \sum_{m = 0}^M\beta_m \sum_{n = 0}^N \mu(A_n \cap B_m)
```
Observe that since $\bigsqcup_{m = 0}^MB_m \supseteq A_n$ and for some $n$ that $\bigsqcup_{m = 0}^M \{A_n \cap B_m\} = A_n$, that $\sum_{m = 0}^M \mu(A_n \cap B_m) = \mu(A_n)$, by countable additivity {prf:ref}`mt:intro:prob_spaces:count_add_setfn`. Then:
```{math}
    \int \varphi + \psi \,\text d \mu &= \sum_{n = 0}^N \alpha_n \mu(A_n) + \sum_{m = 0}^M\beta_m \mu(B_m) \\
    &= \int \varphi \,\text d \mu + \int \psi \,\text d \mu.
```
````

Using only these *three* foundational properties, we will be able to extend the extremely trivial concept of an integral that we outlined in {prf:ref}`mt:int:basics:int:simplefn` to much more complicated functions. Let's move to some corollaries of integration. 

(mt:int:basics:corollaries)=
## Corollaries of integration

If these three properties hold for some class of functions, we obtain (for free) three more desirable results of integration. Notice that these results do *not* make any statements about the functions being simple; they just assume that the functions are in a class of functions where some of the properties we just discussed hold. This means we can pre-prepare some corollaries of integration wherein, if we can define other classes of functions in which those properties *also* hold, these corollaries will extend to those classes of functions, too:

````{prf:corollary} Integration preserves inequalities
:label: mt:int:basics:int:cor:ineq
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. Suppose that $\varphi, \psi : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ are in classes of functions which preserve non-negativity, summation, and rescaling by $-1$ under integration. Then if $\varphi \overset{a.e.}{\leq} \psi$, $\int \varphi\,\text d \mu \leq \int \psi \,\text d \mu$. 
````

````{prf:proof}
Since integration preserves summation and rescaling (here, by $-1$) by supposition, then:
```{math}
\int (\psi - \varphi)\,\text d \mu &= \int \psi \,\text d \mu - \int \varphi \,\text d \mu \\
\Rightarrow \int \psi \,\text d \mu &= \int \varphi \,\text d \mu + \int(\psi - \varphi)\,\text d \mu.
```
Note that if $\varphi \overset{a.e.}{\leq} \psi$, then $\psi - \varphi \overset{a.e.}{\geq} 0$, so $\int (\psi - \varphi) \,\text d \mu \geq 0$, by the non-negativity property. Then:
```{math}
\int \psi \,\text d \mu &= \int \varphi \,\text d \mu + \int(\psi - \varphi)\,\text d \mu\\
\Rightarrow \int \psi\,\text d \mu &\geq \int \varphi \,\text d \mu.
```
````

Using the inequality preservation, we obtain equality preservation, too:

````{prf:corollary} Integration preserves equality
:label: mt:int:basics:int:cor:eq
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. Suppose that $\varphi, \psi : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ are in classes of functions which preserve non-negativity, summation, and rescaling by $-1$ under integration. Then if $\varphi \overset{a.e.}{=} \psi$, $\int \varphi\,\text d \mu = \int \psi \,\text d \mu$. 
````
````{prf:proof}
Note that since $\varphi \overset{a.e.}{=} \psi$, that $\varphi \overset{a.e.}{\leq} \psi$ and $\varphi \overset{a.e.}{\geq} \psi$.

Then by {prf:ref}`mt:int:basics:int:cor:ineq`:
```{math}
    \int \psi\,\text d \mu \geq \int \varphi \,\text d \mu \text{ and } 
    \int \psi\,\text d \mu \leq \int \varphi \,\text d \mu.
```
Together, these two statements imply:
```{math}
    \int \psi\,\text d \mu = \int \varphi \,\text d \mu.
```
````

Finally, the inequality preservation also gives us an intuitive result for absolute values:

````{prf:corollary} Absolute value upper bounds
:label: mt:int:basics:int:cor:abs
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. Suppose that $\varphi : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ is in a class of functions which preserve non-negativity, summation, and rescaling by $-1$ under integration. Then:
```{math}
    \left|\int \varphi \,\text d \mu\right| \leq \int |\varphi| \,\text d \mu.
```
````
````{prf:proof}
By definition of $|\cdot|$ (the absolute value), note that for every $\omega \in \Omega$, $\varphi(\omega) \leq |\varphi(\omega)|$, so $\varphi \leq |\varphi|$ in absolute terms (and consequently, also almost everywhere; e.g., $\varphi \overset{a.e.}{\leq} |\varphi|$, since absolute inequality is stronger than almost everywhere inequality). 

Then by {prf:ref}`mt:int:basics:int:cor:ineq`:
```{math}
    \int \varphi \,\text d \mu \leq \int |\varphi| \,\text d \mu.
```
Further, note that $-\varphi \leq |\varphi|$ (both absolutely and a.e.), so:
```{math}
    \int -\varphi \,\text d \mu &= -\int \varphi \,\text d \mu,\,\,\,\,\text{rescale by $-1$} \\
    &\leq \int |\varphi| \,\text d \mu,
```
where the last line follows, also by {prf:ref}`mt:int:basics:int:cor:ineq`.

Recalling that $|y| = y \vee -y$, then:
```{math}
    \left|\int \varphi \,\text d \mu\right| \leq \int |\varphi| \,\text d \mu.
```
````
Remember that $a \wedge b$ notationally means $\min(a, b)$, and $a \vee b$ notationally means $\max(a, b)$, so $|y| = y \vee -y \equiv \max(y, -y)$ is just another way to write the absolute value. 

We can use these properties of integrals of simple functions, as well as the corollaries, to show that integration can be defined for more complicated functions, too. Let's try bounded functions. 

## Bounded functions

The idea behind a bounded function is basically just that the function has an upper bound. You are probably pretty familiar with this idea from real analysis, so we'll just define it formally here for you:

````{prf:definition} Bounded function
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space and $(\mathbb C, \Sigma)$ is a measurable space. A function $f \in m(\mathcal F, \Sigma)$ is called **bounded** if there exists $M \in \mathbb R$ s.t. for any $\omega \in \Omega$, $|f(\omega)| \leq M$. For such a function, we write $|f| \leq M$. 
````

As it turns out, we can *approximate* the integral of bounded functions using simple functions. Basically, the idea is that we will use simple functions that approximate the bounded function from above, and then simple functions that approximate the bounded function from below. As we make the simple functions sequentially closer and closer approximations (from above and below) of our bounded function, the integrals of the simple functions we are using converge. We can then use the quantity these simple functions converge to as the integral of the bounded function. Since we already have integrals of simple functions, this quantity they converge to (from above and below) is therefore a well-defined quantity (it *exists*), and can be taken to be the integral of the bounded function:

````{prf:theorem} Bounded functions can be approximated by simple functions
:label: mt:int:basics:bounded_int
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. Suppose that $\varphi, \psi : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ are in the class of simple functions, and $f : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ is a bounded function. Suppose further that $E \subseteq \Omega$ is a subset s.t. $\mu(E) < \infty$, and for any $\omega \in E^c$, $f(\omega) = 0$ ($f$ *vanishes* on $E^c$). Then:
```{math}
    \int f \,\text d \mu = \sup_{\varphi \leq f} \int \varphi \,\text d \mu = \inf_{\psi \geq f} \int \psi \,\text d \mu.
```
````
````{prf:proof}
Suppose that $\varphi, \psi$ vanish on $E^c$, and that $f$ is bounded by $M \in \mathbb R$. 

First, we need to show that the sets of these simple functions $\psi : \psi \geq f$ and $\varphi : \varphi \leq f$ are non-empty. Letting $\psi(\omega) = M$ and $\varphi(\omega) = -M$ on $E$ gives the desired result.

Recall that it suffices to show that if $\sup_{\varphi \leq f} \int \varphi \,\text d \mu = \inf_{\psi \geq f}\int \psi \,\text d \mu$, then by the sandwich theorem as $\varphi \leq f \leq \psi$, the desired result holds. 

We will show the desired result in two steps:

1\. $\sup_{\varphi \leq f} \int \varphi \,\text d \mu \leq \inf_{\psi \geq f} \int \psi \,\text d \mu$:

Since $\varphi \leq f \leq \psi$ for all $\varphi, \psi$ that the supremum/infimum are being considered with respect to absolutely, then $\varphi \leq \psi$ absolutely (and hence also a.e., since absolute inequality is stronger than a.e. inequality). 

As $\varphi, \psi$ are simple (and simple functions preserve non-negativity and summation properties of integration), the desired result holds by {prf:ref}`mt:int:basics:int:cor:ineq`.

2\. $\sup_{\varphi \leq f} \int \varphi \,\text d \mu \geq \inf_{\psi \geq f} \int \psi \,\text d \mu$:

For each $k \in \mathbb Z$ where $-n + 1 \leq k \leq n$, let:
```{math}
    E_{n, k} &= \left\{\omega \in E : \frac{kM}{n} \geq f(\omega) > \frac{(k - 1)M}{n}\right\}, \\
    \psi_n(\omega) &= \sum_{k = -n}^n \frac{kM}{n}\mathbb 1_{\{E_{n,k}\}}(\omega) \text{ and }\varphi_n(\omega) = \sum_{k = -n}^n \frac{(k-1)M}{n}\mathbb 1_{\{E_{n,k}\}}(\omega).
```
See {numref}`mt:int:basics:bounding` for details, and note that by construction, $\varphi_n, \psi_n$ are simple functions. By construction, for each $n \in \mathbb N$, that $\bigsqcup_k E_{n,k} = E$, and further, $\psi_n(\omega) - \varphi_n(\omega) = \frac{M}{n}\mathbb 1_{\{E\}}(\omega)$ (since the equality is absolute, it also holds a.e.), so by {prf:ref}`mt:int:basics:int:cor:eq`:
```{math}
    \int (\psi_n - \varphi_n)\,\text d \mu &= \int \frac{M}{n}\mathbb 1_{\{E\}}\,\text d \mu, \\
    &= \frac{M}{n} \mu(E).
```
Further, since simple functions preserve summation by {prf:ref}`mt:int:basics:int:prop:sum_int` and rescaling by $-1$ by {prf:ref}`mt:int:basics:int:prop:rescale`, then:
```{math}
    \int (\psi_n - \varphi_n)\,\text d \mu &= \int \psi_n\,\text d \mu - \int \varphi_n \,\text d \mu = \frac{M}{n} \mu(E) \\
    \Rightarrow\int \varphi_n \,\text d \mu &=  \int \psi_n\,\text d \mu - \frac{M}{n} \mu(E).
```
As $\varphi_n(\omega) \leq f(\omega) \leq \psi_n(\omega)$ by construction, then by {prf:ref}`mt:int:basics:int:prop:sum_int`, and noting that as $\varphi_n \in \{\varphi : \varphi \leq f,\varphi \text{ is simple}\}$ and $\psi_n \in \{\varphi : \varphi \geq f,\psi \text{ is simple}\}$, then for all $n$:
```{math}
    \sup_{\varphi \leq f} \int \varphi \,\text d \mu &\geq \int \varphi_n \,\text d \mu \\
    &= -\frac{M}{n}\mu(E) + \int \psi_n \,\text d \mu \\
    &\geq -\frac{M}{n}\mu(E) + \inf_{\psi \geq f}\int \psi \,\text d \mu \\
    \xrightarrow[n \rightarrow \infty]{} \sup_{\varphi \geq f}\int \varphi \,\text d \mu&\geq \inf_{\psi \geq f}\int \psi \,\text d \mu,
```
which follows because $\mu(E)$ and $M$ are fixed while $n \rightarrow \infty$.

Then by the sandwich theorem since $\varphi \leq f \leq \psi$:
```{math}
    \int f \,\text d \mu = \sup_{\varphi \leq f} \int \varphi \,\text d \mu = \inf_{\psi \geq f} \int \psi \,\text d \mu.
```
````

### Properties of the integral for bounded functions

When functions are bounded on a finite non-vanishing measurable subset of the event space, we can extend all of the properties that we found with simple functions directly to bounded functions. Note that we are going to, in effect, use the fact that we defined the integral as in {prf:ref}`mt:int:basics:bounded_int` as a supremum/infimum of simple functions, and just "borrow" the properties we learned about integrals for simple functions to show the results we want. Let's get started:

````{prf:property} Non-negative bounded function has non-negative integral
:label: mt:int:basics:bounded:prop:nn
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$ is a bounded function that vanishes on a set $E^c$, where $\mu(E) < \infty$. Then if $f \overset{a.e.}{\geq} 0$, $\int f \,\text d \mu \geq 0$.
````

````{prf:proof}
Let $F \in \mathcal F$ be the set on which $f \geq 0$. Let $\varphi_0(\omega) = 0$ for every $\omega \in F$, and let $\varphi_0(\omega) = -M$ for every $\omega \in F^c$, which is a simple function. Since $f \overset{a.e.}{\geq} 0$, then $\mu(F^c) = 0$.

Since $\varphi_0$ is simple, then by {prf:ref}`mt:int:basics:int:simplefn`:
```{math}
    \int \varphi_0\,\text d \mu &= 0\mu(F) + -M\mu(F^c) \\
    &= 0.
```

Note that $\varphi_0 \in \{\varphi : \varphi \leq f, \varphi\text{ is simple}\}$ from the supremum condition {prf:ref}`mt:int:basics:bounded_int`, so:

```{math}
    \int f \,\text d \mu &=  \sup_{\varphi \leq f}\int \varphi \,\text d\mu\\
    &\geq \int \varphi_0 \,\text d\mu = 0,
```
because the supremum might be over other simple functions as well.
````

So what we did here was, we defined a simple function (the *zero* function) for the entire part of the sample space on which $f \geq 0$, and set its other values using the bound on $f$ to definitively obtain a simple function that is $\leq f$. Then, we used that the supremum would be at least as large as the integral of one of the member functions to establish the appropriate bound.

````{prf:property} Rescaled bounded function has rescaled integral
:label: mt:int:basics:bounded:prop:rescale
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ is a bounded function that vanishes on a set $E^c$, where $\mu(E) < \infty$. Then if $r \in \mathbb R$, $\int r f \,\text d \mu = r\int f \,\text d \mu$.
````

````{prf:proof}
Suppose WOLOG that $r > 0$. Then $r \varphi \leq r f \iff \varphi \leq f$, and consequently by {prf:ref}`mt:int:basics:bounded_int`:

```{math}
    \int rf \,\text d \mu &= \sup_{\varphi \leq f}\int r\varphi \,\text d \mu \\
    &= \sup_{\varphi \leq f }r \int \varphi \,\text d \mu,
```
which is because $\varphi$ are simple, and by {prf:ref}`mt:int:basics:int:prop:rescale`. Then:
```{math}
    \int rf \,\text d \mu &= \sup_{\varphi \leq f }r \int \varphi \,\text d \mu \\
    &=  r\sup_{\varphi \leq f } \int \varphi \,\text d \mu \\
    &= r\int f \,\text d \mu,
```
which is by {prf:ref}`mt:int:basics:bounded_int`.
````


````{prf:property} Summed bounded functions have summed integrals
:label: mt:int:basics:bounded:prop:sum
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f,g : (\Omega, \mathcal F) \rightarrow (\mathbb R, \Sigma)$ are bounded functions that vanishes on a set $E^c$, where $\mu(E) < \infty$. Then $\int (f + g)\,\text d \mu = \int f \,\text d \mu + \int g \,\text d \mu$.
````

````{prf:proof}
Note that if $\psi_1 \geq f$ and $\psi_2 \geq g$, then $\psi_1 + \psi_2 \geq f + g$. Further, note that $\{\psi : \psi \geq f + g\} \subseteq \{ \psi_1 + \psi_2 : \psi_1 \geq f, \psi_2 \geq g\}$. Consequently:
```{math}
    \int (f + g) \,\text d \mu &= \inf_{\psi \geq f + g}\int \psi \,\text d \mu \\
    &\leq \inf\left\{\int (\psi_1 + \psi_2) \,\text d \mu : \psi_1 \geq f, \psi_2 \geq g\right\} \\
    &= \inf\left\{\int \psi_1 \,\text d \mu + \int \psi_2 \,\text d \mu : \psi_1 \geq f, \psi_2 \geq g\right\},
```
which is by {prf:ref}`mt:int:basics:int:prop:sum_int` since $\psi_1, \psi_2$ are simple. Continuing:
```{math}
:label: mt:int:basics:bounded:prop:sum:eqhelp1
    \int (f + g) \,\text d \mu &\leq \inf\left\{\int \psi_1 \,\text d \mu + \int \psi_2 \,\text d \mu : \psi_1 \geq f, \psi_2 \geq g\right\} \\
    &= \inf_{\psi_1 \geq f}\int \psi_1 \,\text d \mu  + \inf_{\psi_2 \geq g}\int \psi_2 \,\text d \mu \\
    &= \int f \,\text d \mu + \int g \,\text d \mu.
```
Repeating the same argument on $-f$ and $-g$, we obtain:
```{math}
:label: mt:int:basics:bounded:prop:sum:eqhelp2
    \int -(f + g) \,\text d \mu \leq -\int f \,\text d \mu - \int g \,\text d \mu \\
    \Rightarrow \int (f + g) \,\text d \mu \geq \int f \,\text d \mu + \int g \,\text d \mu,
```
which is by rescaling by $-1$ from {prf:ref}`mt:int:basics:bounded:prop:rescale`. Combining {eq}`mt:int:basics:bounded:prop:sum:eqhelp1` and {eq}`mt:int:basics:bounded:prop:sum:eqhelp2`:
```{math}
\int (f + g) \,\text d \mu \geq \int f \,\text d \mu + \int g \,\text d \mu.
```
````

Notice that since bounded functions obey inequalities with respect to $0$, are rescalable by $-1$, and their sums have summed integrals, that we obtain all of the integration corollaries in {numref}`mt:int:basics:corollaries` for free!

## Non-negative functions

Next, we're going to extend our repetoire of functions with well-defined integrals to the non-negative functions:

````{prf:definition} Non-negative function
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measure space. A function $f \in m(\mathcal F, \Sigma)$ is called **non-negative** if for every $\omega \in \Omega$, $f(\omega) \geq 0$. If $f$ is a non-negative function, we write $f \geq 0$.
````

To begin to define the integral of a non-negative function, we need to be able to integrate over sets:

````{prf:definition} Integral over a set
:label: mt:int:basics:set_int
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$. The integral over a set $E \in \mathcal F$ is:
```{math}
    \int_E f \,\text d \mu \triangleq \int f \cdot \mathbb 1_{\{E\}}\,\text d \mu.
```
````

Now, we are ready to define the integral of a non-negative function:

````{prf:definition} Integral of a non-negative function
:label: mt:int:basics:nn:int_def
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measure space, and a function $f \in m(\mathcal F, \Sigma)$ is s.t. $f \geq 0$. Then:
```{math}
    \int f \,\text d \mu \triangleq \sup \left\{\int h \,\text d \mu : 0 \leq h \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : h(\omega) > 0\}\right) < \infty\right\}.
```
````
As we can see, this quantity is clearly well-defined (it contains at least one element): simply take $h(\omega) = 0$ for all $\omega \in \Omega$. Further, it is the supremum of the integrals of bounded functions, so it also could be non-finite. To calculate the integral of a bounded function, remember we just introduced a method of computing the integral from simple functions which approximated $f$ from above and below. As we made these coarse simple functions more fine, their integrals ended up converging, so we ended up being able to sandwich the integral of our bounded function between them. 

Similarly, we also need an approach to calculate the integral for non-negative functions, so we'll introduce this lemma:

````{prf:lemma} Computing integral of non-negative function
:label: mt:int:basics:nn:int_compute
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measure space, and a function $f \in m(\mathcal F, \Sigma)$ is s.t. $f \geq 0$. Then if $(F_n) \subseteq \mathcal F$ is s.t. $F_n \uparrow \Omega$ and $\mu(F_n) < \infty$ for all $n \in \mathbb N$, then:
```{math}
    \int_{F_n} (f \wedge n) \,\text d \mu \uparrow \int f \,\text d \mu\text{ as }n \rightarrow \infty.
```
````
Remember that $f \wedge n$ means $\min(f, n)$. So, in effect here, what we are doing is "chopping off" $f$ at a suitable upper bound $n$ along a $\mu$-finite subset of the $\sigma$-algebra that is converging (from below) to the event space. The non-negativity assumption of $f$ suggests that this quantity "should" keep growing; it's pretty obvious that $\int_{F_n} f \wedge n \,\text d \mu \leq  \int f \,\text d \mu$. The problem is that, along this convergent sequence of sets, we could end up with an accumulation point *below* the value we are interested in, $\int f \,\text d \mu$. The purpose of this lemma is to establish that the desired quantity is, in fact, a *limit* and not just an upper bound.

````{prf:proof}
Let $h_n \triangleq (f \wedge n)\mathbb 1_{\{F_n\}}$, and suppose that $M_n$ is an upper bound for $h_n$. Note that this upper bound clearly exists, as by construction, $h_n \leq n$. 

It is clear that for each $h_n \in \left\{0 \leq h \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : h(\omega) > 0\}\right) < \infty\right\}$, so each $h_n$ is a possibility for $h$ in the supremum. 

Further, note that $\int h_n \,\text d \mu$ is a monotone non-decreasing subset, as for every $n$:
```{math}
    \int h_n \,\text d \mu &= \int_{F_n}(f \wedge n) \,\text d \mu \\
    &\leq \int_{F_n}(f \wedge n + 1) \,\text d \mu + \int_{F_{n + 1} \setminus F_n}(f \wedge n + 1) \,\text d \mu,\,\,\,\,\int_{F_{n + 1} \setminus F_n}(f \wedge n + 1) \,\text d \mu \geq 0 \\
    &= \int_{F_{n + 1}}(f \wedge n + 1) \,\text d \mu \\
    &= \int h_n \,\text d \mu
```
Which follows because $(n + 1) \geq n$, $F_n \subseteq F_{n + 1}$, and $(f \wedge n + 1) \geq 0$ is upper-bounded by $M_n$ so {prf:ref}`mt:int:basics:bounded:prop:nn` gives that the integral is non-negative.

Suppose that $h \in \left\{0 \leq h \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : h(\omega) > 0\}\right) < \infty\right\}$ is an arbitrary element of the supremum set with an upper bound $M$. Then for any $n \geq M$, we obtain:
```{math}
    \int_{F_n} f \wedge n \,\text d \mu &= \int h_n \,\text d \mu \equiv \int_{F_n} h_n \,\text d \mu \\
    &\geq \int_{F_n} h \,\text d \mu,\,\,\,\,0 \leq h \leq f, h \leq M \Rightarrow h \leq h_n
```
The bottom line follows because $h_n = f \wedge n$, and $n \geq M$, so $h_n \geq f \wedge M$. Further, $h$ is upper bounded by $M$ and at most the value of $f$, so $h \leq f \wedge M \leq h_n$. Then since $h \leq h_n$, it follows by {prf:ref}`mt:int:basics:int:cor:ineq` that $\int_{F_n} h \,\text d \mu \leq \int_{F_n} h_n \,\text d \mu$. Continuing and applying {prf:ref}`mt:int:basics:bounded:prop:sum`:
```{math}
    :label: mt:int:basics:nn:int_def:helper1
    \int_{F_n} f \wedge n \,\text d \mu \geq \int h \,\text d \mu - \int_{F_n^c} h \,\text d \mu.
```
Which follows by noting that $h = h \mathbb 1_{\{F_n\}} + h\mathbb 1_{\{F_n^c\}}$, and using the definition of the integral over a set.

Noting that $0 \leq h \leq M$, then by {prf:ref}`mt:int:basics:int:cor:ineq`:
```{math}
    \int_{F_n^c}0 \,\text d \mu &\leq \int_{F_n^c}h \,\text d \mu \leq \int_{F_n^c}M \,\text d \mu \\
    \Rightarrow 0 &\leq \int_{F_n^c}h \,\text d \mu \leq M\mu\left(F_n^c \right) \downarrow 0\text{ as }n \rightarrow \infty \\
    \Rightarrow \int_{F_n^c}h \,\text d \mu &\xrightarrow[n \rightarrow \infty]{} 0,\,\,\,\,\text{ Sandwich Theorem}
```
Which follows because $F_n \uparrow \Omega$ implies that $F_n^c \downarrow \varnothing$, so by convergence from above {prf:ref}`mt:intro:prob_spaces:meas:convabove`, $\mu(F_n^c) \downarrow 0$. 

Therefore combining this with {eq}`mt:int:basics:nn:int_def:helper1` gives that for every $h \in  \left\{0 \leq h \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : h(\omega) > 0\}\right) < \infty\right\}$:
```{math}
    \liminf_{n \rightarrow \infty} \int_{F_n}(f \wedge n) \,\text d \mu \geq \int h \,\text d \mu.
```
````

### Properties of the integral for non-negative functions

At this point, you are probably becoming pretty accustomed to the procedures we are going here. Since we just defined the integral of a non-negative function in terms of bounded functions in {prf:ref}`mt:int:basics:nn:int_def`, we're going to use properties about integrals of bounded functions to prove properties about the integrals of non-negative functions. Let's get started:

````{prf:property} Non-negative function has non-negative integral
:label: mt:int:basics:nn:nn
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$ is s.t. $f \geq 0$. Then $\int f \text d \mu \geq 0$.
````

````{prf:proof}
Take $h(\omega) = 0$ for all $\omega \in \Omega$. Note that $h \in \left\{h : 0 \leq h \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : h(\omega) > 0\}\right) < \infty\right\}$.

Further, by {prf:ref}`mt:int:basics:int:simplefn`, $\int h\,\text d \mu = 0$. 

Then $\int f \,\text d \mu \geq \int h\,\text d \mu = 0$, since $\int f \, \text d\mu$ is a supremum over elements including $0$. 
````

Well, that one was pretty obvious, let's try rescaling:

````{prf:property} Rescaled non-negative function has rescaled integral
:label: mt:int:basics:nn:rescale
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$ is s.t. $f \geq 0$. Then $\int rf \,\text d \mu = r\int f \,\text d \mu$.
````
````{prf:proof}
If $r = 0$, the result is immediate, so we'll consider the case where $r \neq 0$.

WOLOG suppose that $r > 0$. 

Then using {prf:ref}`mt:int:basics:nn:int_def`, for $r > 0$:
```{math}
    \int rf \,\text d \mu &= \sup \left\{\int q \,\text d \mu : 0 \leq q \leq rf, q \text{ is bounded}, \mu\left(\{\omega \in \Omega : q(\omega) > 0\}\right) < \infty\right\} \\
    &= \sup \left\{\int rh \,\text d \mu : 0 \leq \frac{h}{r} \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : q(\omega) > 0\}\right) < \infty\right\}
```
Notice that:
1. $q \leq rf \iff \frac{q}{r} \leq f$ for $r > 0$,
2. that $\{q(\omega) > 0\} = \{\frac{q}{r}(\omega) > 0\} \Rightarrow \mu(q > 0) = \mu\left(\frac{q}{r} > 0\right)$, and 
3. $q$ is bounded by $M$ $\iff$ $\frac{q}{r}$ is bounded by $\frac{M}{r}$. 

Letting $h \triangleq \frac{q}{r}$ gives:
```{math}
    \int rf \,\text d \mu &= \sup \left\{\int rh \,\text d \mu : 0 \leq h \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : h(\omega) > 0\}\right) < \infty\right\} \\
    &= \sup \left\{r\int h \,\text d \mu : 0 \leq h \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : h(\omega) > 0\}\right) < \infty\right\},
```
where we exchanged the integral of the rescaling with the rescaled integral by {prf:ref}`mt:int:basics:bounded:prop:rescale`. Continuing:
```{math}
    \int rf \,\text d \mu &= r\sup \left\{\int h \,\text d \mu : 0 \leq h \leq f, h \text{ is bounded}, \mu\left(\{\omega \in \Omega : h(\omega) > 0\}\right) < \infty\right\} \\
    &= r\int f \,\text d \mu.
```
````
This one is also pretty obvious, and basically come down to the fact that we were taking a supremum over what was, effectively, the same set (just rescaled) due to the equivalences and $\iff$ statements, and then using that $r > 0$ implied that the supremum of the rescaled values was the rescaling of the supremum of the original values. We could repeat this proof with $r < 0$ to obtain that the result holds for *any* $r$, with suitable adjustments of the inequalities to accomodate the negative values (hence, the WOLOG). 

Now, our final aim is summation:

````{prf:property} Summed non-negative functions have summed integrals
:label: mt:int:basics:nn:sum
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f, g \in m(\mathcal F, \Sigma)$ are s.t. $f, g \geq 0$. Then $\int (f + g)\,\text d \mu = \int f \,\text d \mu + \int g \,\text d \mu$.
````
We'll address this by showing that Left $\leq$ RHS, and RHS $\leq$ LHS, and therefore RHS $=$ LHS:
````{prf:proof}
1\. $\int (f + g)\,\text d \mu \leq \int f \,\text d \mu + \int g \,\text d \mu$:

Note that $(a + b)\wedge n \leq (a \wedge n) + (b \wedge n)$.

Let $(F_n) \subseteq \mathcal F$ be a $\mu$-finite sequence where $F_n \uparrow \Omega$ from below. Then:
```{math}
    \int_{F_n} (f + g)\wedge n \,\text d \mu &\leq \int_{F_n} \left[(f \wedge n) + (g \wedge n)\right]\,\text d \mu \\
    &= \int_{F_n} (f \wedge n)\,\text d \mu + \int_{F_n}(g \wedge n)\,\text d \mu,
```
which is because $(f \wedge n) \mathbb 1_{F_n}$ and $(g \wedge n) \mathbb 1_{F_n}$ are bounded, so application of {prf:ref}`mt:int:basics:bounded:prop:sum` gives the desired result. Then by {prf:ref}`mt:int:basics:nn:int_compute`:
```{math}
\int_{F_n} (f + g)\wedge n \,\text d \mu &\leq \int_{F_n} (f \wedge n)\,\text d \mu + \int_{F_n}(g \wedge n)\,\text d \mu \\
&\uparrow \int f\,\text d \mu + \int g\,\text d \mu.
```

2\. $\int (f + g)\,\text d \mu \geq \int f \,\text d \mu + \int g \,\text d \mu$:

Suppose that $h \in \{h : 0 \leq h \leq f,h \text{ is bounded}, \mu(h > 0) < \infty\}$, and that $k \in \{k : 0 \leq k \leq g,k \text{ is bounded}, \mu(k > 0) < \infty\}$. Then by definition, it follows that $h \leq f, k \leq g$, so $h + k \leq f + g$.

Therefore:
```{math}
&\left\{q : 0 \leq q \leq f + g, q \text{ is bounded}, \mu\left(q > 0\right) < \infty\right\} \supseteq \\
&\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\bigg\{\int (h + k) \,\text d \mu : 0 \leq h \leq f, 0 \leq k \leq g, h\text{ and }k \text{ are bounded},\\
&\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\mu\left(q > 0\right), \mu(k > 0) < \infty\bigg\}.
```

So:
```{math}
    \int (f + g) \,\text d \mu &= \sup \left\{\int q \,\text d \mu : 0 \leq q \leq f + g, q \text{ is bounded}, \mu\left(q > 0\right) < \infty\right\} \\
    &\geq \sup\bigg\{\int (h + k) \,\text d \mu : 0 \leq h \leq f, 0 \leq k \leq g, h\text{ and }k \text{ are bounded}, \\
    &\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\mu\left(q > 0\right), \mu(k > 0) < \infty\bigg\} \\
    &=  \sup \left\{\int h \,\text d \mu : 0 \leq h \leq f, h \text{ is bounded}, \mu(h > 0) < \infty\right\} + \\
    & \;\;\;\;\;\;\;\;\;\sup \left\{\int k \,\text d \mu : 0 \leq k \leq g, k \text{ is bounded}, \mu(k > 0) < \infty\right\} \\
    &= \int f \,\text d \mu + \int g \,\text d \mu,
```
which follows by the {prf:ref}`mt:int:basics:nn:int_def`, the integral of a non-negative function.
````

As before, since non-negative functions obey inequalities with respect to $0$, are rescalable by $-1$, and their sums have summed integrals, we obtain all of the integration corollaries in {numref}`mt:int:basics:corollaries`. 

We are now ready to relax the class of functions we can integrate to the broadest class of functions we will deal with in this course: the integrable functions.

## $\mu$-integrable functions

Let's start with a quick definition:

````{prf:definition} $\mu$-integrable functions
:label: mt:int:basics:muint_fn
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, and $(\mathbb R, \Sigma)$ is a measurable space. We say that $f \in m(\mathcal F, \Sigma)$ is $\pmb\mu$-**integrable** if $\int |f| \,\text d \mu < \infty$.
````

Like before, we'll now start with a definition of an integral for a $\mu$-integrable function:

````{prf:definition} Integral of a $\mu$-integrable function
:label: mt:int:basics:muint:int
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$ is $\mu$-integrable, and for each $\omega \in \Omega$, define:
```{math}
    f^+(\omega) \triangleq f(\omega) \wedge 0 \text{ and }f^-(\omega) \triangleq (-f(\omega)) \wedge 0.
```
Then the integral is:
```{math}
    \int f \,\text d \mu = \int f^+ \,\text d \mu - \int f^- \,\text d \mu.
```
````

It should be pretty immediately clear to you that for every $\omega \in \Omega$, that $f(\omega) = f^+(\omega) - f^-(\omega)$, and that $|f(\omega)| = f^+(\omega) + f^-(\omega)$, just by construction. Another pretty easy interesting note to conclude that $f^+(\omega), f^-(\omega) \leq |f(\omega)|$. 

Since $f^+ \geq 0$ and $f^- \geq 0$, by {prf:ref}`mt:int:basics:nn:int_def` and {prf:ref}`mt:int:basics:int:cor:ineq` if $f$ is $\mu$-integrable:
```{math}
:label: mt:int:basics:muint:helper
    \int f^+ \,\text d \mu,\int f^- \,\text d \mu < \int |f| \,\text d \mu < \infty
```
Just by looking off at the {prf:ref}`mt:int:basics:muint_fn` for a $\mu$-integrable function, which also directly that the quantity is well-defined.

### Properties of $\mu$-integrable functions

Well, this time we know that we can compute the integral using {eq}`mt:int:basics:muint:compute`, as well as the corresponding discussion that we had. But we still need some more machinery to show that our basic properties of integration hold like before:

````{prf:lemma} Integral of a difference
:label: mt:int:basics:muint:intdiff
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$ is $\mu$-integrable, where $f = f_1 - f_2$, and $f_1, f_2 \geq 0$, where for each $i \in \{1, 2\}$, $\int f_i \,\text d \mu < \infty$. Then:
```{math}
    \int f \,\text d \mu = \int f_1 \, \text d \mu - \int f_2 \,\text d \mu.
```
````
````{prf:proof}
Note that with $f^+, f^-$ as defined above, that $f = f^+ - f^- = f_1 - f_2$. Then:
```{math}
    f_1 + f^- = f_2 + f^+
```
where $f^+, f^-, f_1, f_2 \geq 0$.

Then by repeatedly applying {prf:ref}`mt:int:basics:int:cor:eq` for non-negative functions and {prf:ref}`mt:int:basics:nn:sum`:
```{math}
    \int f_1 \text d \mu + \int f^- \text d \mu &= \int f_1 + f^- \text d \mu \\
    &= \int f_2 + f^+ \text d \mu \\
    &= \int f_2 \text d \mu + \int f^+ \text d \mu \\
    \Rightarrow \int f_1 \text d \mu - \int f_2 \text d \mu &= \int f^+ \text d \mu - \int f^- \text d \mu \\
    &= \int f \text d \mu,
```
Since RHS is the definition of a $\mu$-integrable function's integral, {prf:ref}`mt:int:basics:muint:int`.
````

And now, onto our properties. This first one seems *extremely* obvious, but as it turns out, it requires a little bit of nuance. We'll introduce a general purpose lemma here, which will let us handle this in the future:

````{prf:lemma} Integral of non-negative function on a $\mu$-null set is $0$
:label: mt:int:basics:muint:munull:nn
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$ is non-negative. Then if $F$ is a $\mu$-null set, $\int_{F} f \,\text d \mu = 0$.
````
````{prf:proof}
Recall that $\int_{\{F\}} f \,\text d \mu \triangleq \int f\mathbb 1_{\{F\}} \,\text d \mu$, by {prf:ref}`mt:int:basics:set_int`.

With $h \triangleq f \mathbb 1_{\{F\}}$, note that the integral of $f$ itself is defined as a supremum over bounded functions by {prf:ref}`mt:int:basics:nn:int_def`, which are defined as a supremum over simple functions by {prf:ref}`mt:int:basics:bounded_int`. 

Suppose that $\varphi$ is any such simple function, and let $\{(\alpha_n, E_n)\}_{n \in [N]} \subseteq \mathcal F$ be the coefficients/disjoint sets with which each are defined, by {prf:ref}`mt:int:basics:int:simplefn`. 

Note that with $h \triangleq f \mathbb 1_{\{F\}}$, that the simple functions which are used to define the integral of $h$ will be $\varphi \mathbb 1_{\{F\}}$, and will have coefficients/disjoint sets $\{(\alpha_n, E_n \cap F)\}_{n \in [N]}$. However, for every such simple function:
```{math}
    \int \varphi \mathbb 1_{\{F\}}\text d \mu &= \sum_{n \in [N]}\alpha_n \mu(E_n \cap F) \\
    &= 0\sum_{n \in [N]}\alpha_n = 0,
```
because $0 \leq \mu(E_n \cap F) \leq \mu(F) = 0$.

Working back upwards, all of the simple functions, all of the bounded functions, and consequently $h$ will have an integral $0$. 
````
And we can extend it extremely trivially to $\mu$-integrable functions, too (try it!):
````{prf:corollary} Integral of $\mu$-integrable function on a $\mu$-null set is $0$
:label: mt:int:basics:muint:munull:muint
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$ is $\mu$-integrable. Then if $F$ is a $\mu$-null set, $\int_{F} f \,\text d \mu = 0$.
````
Since bounded and simple functions are special cases of $\mu$-integrable functions, the machinery will readily apply to those cases, too.

Now, we are ready to handle the properties:


````{prf:property} Non-negative $\mu$-integrable function has non-negative integral
:label: mt:int:basics:muint:nnint
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, and $f \in m(\mathcal F, \Sigma)$ is $\mu$-integrable where $f \overset{a.e.}{\geq} 0$. Then $\int f \,\text d \mu \geq 0$.
````
````{prf:proof}
Let $F^c$ be the $\mu$-null set on which $f(\omega) < 0$, and define $f^+$ and $f^-$ as in {prf:ref}`mt:int:basics:muint:int`. By construction, note that $(f^-) \mathbb 1_{\{F^c\}} \equiv f^- \geq 0$. Then by {prf:ref}`mt:int:basics:muint:munull:nn`:
```{math}
    \int f^- \,\text d \mu &= \int f^- \mathbb 1_{\{F^c\}}\,\text{d} \mu \\
    &= 0.
```

Then:
```{math}
    \int f \,\text d \mu &= \int f^+ \,\text d \mu - \int f^- \,\text d \mu \\
    &= \int f^+ \,\text d \mu \geq 0,
```
which is by {prf:ref}`mt:int:basics:nn:nn` since $f^+ \geq 0$.
````

We can rescale them:
````{prf:property} Rescaled $\mu$-integrable functions have rescaled integrals
:label: mt:int:basics:muint:rescale
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, $f \in m(\mathcal F, \Sigma)$ is $\mu$-integrable, and $r \in \mathbb R$. Then $\int r f \,\text d \mu = r \int f \,\text d \mu$.
````
````{prf:proof}
Define $f^+$ and $f^-$ as in {prf:ref}`mt:int:basics:muint:int`. Note that since $f^+$ and $f^-$ are non-negative, then by {prf:ref}`mt:int:basics:nn:sum`:
```{math}
    \int rf \,\text d \mu &= \int r(f^+ - f^-) \,\text d \mu \\
    &= \int rf^+ \,\text d \mu - \int rf^- \,\text d \mu \\
    &= r\left(\int rf^+ \,\text d \mu - \int f^- \,\text d \mu\right) \\
    &= r \int f \,\text d \mu,
```
where the second-to-last line follows by {prf:ref}`mt:int:basics:nn:rescale`. 
````

and finally, we can sum them:
````{prf:property} Summed $\mu$-integrable functions have summed integral
:label: mt:int:basics:muint:sum
Suppose that $(\Omega, \mathcal F, \mu)$ is a measure space, $(\mathbb R, \Sigma)$ is a measurable space, $f,g \in m(\mathcal F, \Sigma)$ is $\mu$-integrable. Then $\int (f + g)\,\text d \mu = \int f \,\text d \mu + \int g \,\text d \mu$.
````
````{prf:proof}
Define $f^+, f^-, g^+, g^-$ as in {prf:ref}`mt:int:basics:muint:int`. Note that:
```{math}
    f + g &= (f^+ - f^-) + (g^+ - g^-) \\
    &= (f^+ + g^+) - (f^- + g^-).
```
Where $(f^+ + g^+), (f^- + g^-) \geq 0$, and further, have finite integrals since $f, g$ are $\mu$-integrable, the sum of non-negative functions are non-negative, and then application of {eq}`mt:int:basics:muint:helper`. 

Note further that $|f + g| \leq |f| + |g|$ by the triangle inequality, so $f, g$ $\mu$-integrable implies that $f + g$ is $\mu$-integrable. Then by {prf:ref}`mt:int:basics:muint:intdiff`:
```{math}
    \int(f + g)\,\text d \mu &= \int (f^+ + g^+)\,\text d\mu - \int (f^- + g^-)\,\text d\mu \\
    &= \int f^+\,\text d\mu + \int g^+\,\text d\mu - \left(\int f^-\,\text d\mu + \int g^-\,\text d\mu\right),
```
which is because $f^+, g^+, f^-, g^- \geq 0$ by {prf:ref}`mt:int:basics:nn:sum`. Simply rearranging terms:
```{math}
    \int(f + g)\,\text d \mu &= \int f^+\,\text d\mu - \int f^-\,\text d\mu + \int g^+\,\text d\mu - \int g^-\,\text d\mu \\
    &= \int f \,\text d \mu + \int g \,\text d \mu,
```
since these are their definitions given by {prf:ref}`mt:int:basics:muint:int`.
````

As always, we obtain all of the integration corollaries in {numref}`mt:int:basics:corollaries` due to these three properties holding for $\mu$-integrable functions.

````{prf:remark} Constant functions follow all the properties discussed
:label: mt:int:basics:constants
Suppose that you have an extremely simple function; e.g., $f(\omega) = x$, where $x \in \mathbb R$ is a constant. This function is certainly simple; e.g., take $\alpha_n = x$, $F_1 = \Omega$. It should be pretty clear to you that because it is simple, it is obviously bounded (you can use the number $x$ itself as the bound), it is a rescaling of a non-negative function (if $x < 0$, consider that $f(\omega) = -g(\omega)$, where $g(\omega) = -x \geq 0$), and it is therefore integrable too, assuming that the measure of the whole space is finite.
````
Think about this remark a little bit.
